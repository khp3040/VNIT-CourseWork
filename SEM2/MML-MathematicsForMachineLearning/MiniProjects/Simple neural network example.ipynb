#Simple neural network example 

###First, we load the necessary libraries

```
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

###Then, we create a simple neural network

```
model=keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.add(keras.layers.Dense(1))
```

###Next, compile the model

```
model.compile(optimizer='sgd', loss='mean_squared_error')
```

###Create 
1. input data, variable x
2. target data, variable y

Note that: $y=2x-1$

```
x=np.array([-1, 0,1,2,3,4],dtype=float)
y=np.array([-3,-1,1,3,5,7],dtype=float)
```

###Train model

```
model.fit(x,y,epochs=500)
```

###Use model to predict responce with $x=3.5$ as input.
###Based on formula $y=2x-1$ the predicted value should be close to 6.

```
print(model.predict([3.5]))
```

###We can view a summary of the model 

```
model.summary()
```
